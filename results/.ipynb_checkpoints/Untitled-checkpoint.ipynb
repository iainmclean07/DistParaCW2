```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

go_medians_DS1 = [16.162, 8, 4.18, 2.186, 1.12, 0.743, 0.750, 0.736]
c_medians_DS1 = [15.288, 7.661, 4.092, 2.097, 1.073, 0.688, 0.678, 0.674]
erl_medians_DS1 = [28.436, 21.650, 13.276, 7.374, 3.836, 2.315, 1.957, 1.830]

go_medians_DS2 = [68.391, 34.313, 17.813, 9.232, 4.651, 3.093, 3.083, 3.078]
c_medians_DS2 = [65.638, 33.104, 17.163, 8.937, 4.514, 2.829, 2.812, 2.808]
erl_medians_DS2 = [119.989, 91.355, 55.722, 30.922, 16.426, 9.584, 7.824, 7.245]

go_medians_DS3 = [39.251, 19.675, 12.916, 12.909, 12.909]
c_medians_DS3 = [38.054, 19.118, 11.897, 11.850, 11.854]
erl_medians_DS3 = [132.396, 70.274, 41.123, 33.811, 30.872]
```

```python
figure(figsize=(10, 10))

N = 8

ind = np.arange(N) 
width = 0.35       
plt.plot(ind, go_medians_DS1, label='Go')
plt.plot(ind, c_medians_DS1, label='C')
plt.plot(ind, erl_medians_DS1, label='Erlang')

plt.ylabel('Runtime (s)')
plt.xlabel('Threads')
plt.title('DS1 Runtimes')

plt.xticks(ind + width / 2, ('1', '2', '4', '8', '16', '32', '48', '64'))
plt.legend(loc='best')
plt.show()
```

```python
figure(figsize=(10, 10))

c_t1s = 15.795
go_t1s = 15.963
erl_t1s = 28.914

labels = [1, 2, 4, 8, 16, 32, 48, 64]

c_absolute_speedups = [c_t1s/x for x in c_medians_DS1]
go_absolute_speedups = [go_t1s/x for x in go_medians_DS1]
erl_absolute_speedups = [erl_t1s/x for x in erl_medians_DS1]

N = 8

ind = np.arange(N) 
width = 0.27       
plt.plot(ind, c_absolute_speedups, label='C')
plt.plot(ind, go_absolute_speedups, label='Go')
plt.plot(ind, erl_absolute_speedups, label='Erlang')
plt.plot(ind, labels, label='Ideal')

plt.ylabel('Speedup')
plt.xlabel('Threads')
plt.title('DS1 Absolute Speedup')

plt.xticks(ind + width / 2, labels)
plt.legend(loc='best')
plt.show()
```
